```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")
```

```python
fmri_avg = fmri.groupby(["timepoint", "region", "event"])["signal"].mean().reset_index()
sns.set_theme(style = "whitegrid")
sns.FacetGrid(fmri_avg , height = 3 , aspect =2)
sns.lineplot(data=fmri_avg, x="timepoint", y="signal", hue="region", style="event", markers=True)
plt.title("Trend of Brain Activity Over Time")
```

```python
sns.displot(fmri, x="signal", hue="region", col="event", kind="kde", height=4, aspect=1)
```

```python
sns.boxplot(data=fmri, x="event", y="signal", hue="region")
plt.title("Signal Distribution by Event and Region")
```

```python
pivot = fmri.pivot_table(values="signal", index="timepoint", columns="region", aggfunc="mean")
sns.heatmap(pivot, cmap="coolwarm", annot=False)
```
